```python
# importing pandas
import pandas as pd

# read csv file
df=pd.read_csv("nacgfs-jun-2023.csv")

# head tag used to get first 5 rows of the data set by deafult
df.head()
```

```python
# INDEXING AND SLICING

# LOC :  Loc is used to filter rows and select columns by lable[Rows-"Row's Index", column-"Column's Name"]

df.loc[0,"Series_reference"]#returns value at index 0 of column Series_reference 
#here slicing is being done for 0 index  Series_reference

#lets see now multiple data fetch using loc
df.loc[[0,1,2],:]#at row part we gave index 0,1,2 and at column we gave : which means it will return all data for first 3 rows
#OR
df.loc[0:3,:]
```

```python
# ILOC : iloc is used to filter rows and columns based on integer position"

# df.iloc[index_of_rows,index_of_column]

df.iloc[0:4,[0,1,3]]

# here the row index are list containing(0,1,2,4) and index of column 0,1,3
```

```python
# columns contains in data frames 

df.columns
```

```python
# Boolean Indexing

mask = df['Period'] >= 2015

# We can then use this mask to filter the DataFrame using the loc attribute:
df.loc[mask]
```

```python
# we can create a boolean mask using the str.contains method:
mask = df.columns.str.contains('Type')
'''This creates a boolean mask with the same length as the number of columns, 
where each value is True if the corresponding column label contains the word "Type", and False otherwise.'''
```

```python
# we can see column name which has type keyowrd in it

df.loc[:,mask]
```

```python
# Data Manipulation 
# Reshaping:

# Pivoting: pivot(): Rotates data from rows into columns based on a key column.
df.pivot(index=['Series_reference', 'Period'], columns='STATUS', values='Data_value')
```

```python
# Melting: melt(): Converts data from wide format to long format.
pd.melt(df, id_vars=['Series_reference', 'Period'], var_name='STATUS')
```

```python
# Stacking/Unstacking: 
# stack(): Collapses DataFrame columns into a single column.
# unstack(): Reverse operation of stacking.

df.set_index(['Series_reference', 'Period'])[['Data_value', 'STATUS', 'Group']].stack().reset_index()
```

```python
# Reshaping with pivot_table:

# pivot_table(): Similar to pivot, but can handle duplicate entries by applying an aggregation function.

df.pivot_table(index='Series_reference', columns='Period', values='Data_value', aggfunc='mean')
```

```python
# Transforming:

# Grouping and Aggregation:

# groupby(): Groups data based on one or more keys.
df.groupby(['Series_reference', 'Period'])['Data_value'].mean().reset_index()
```

```python
# Dropping Duplicates:

# drop_duplicates(): Removes duplicate rows based on a subset of columns or all columns.

deduped_df = df.drop_duplicates(subset=['Series_reference', 'Period'])
deduped_df
```

```python
# Renaming Columns:

# rename(): Renames columns to make them more readable or consistent.

renamed_df = df.rename(columns={'Series_reference': 'Series', 'Data_value': 'Values'})
renamed_df.columns
```
